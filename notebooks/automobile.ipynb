```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
auto_data = pd.read_csv("Automobile_data.csv")
auto_data.head(10)
```

```python
print(auto_data.describe())
# auto_data['symboling'].unique()
print(auto_data.shape)
print(auto_data.info())
```

```python
## checking for null values
print(auto_data.isnull().sum())
print(auto_data[auto_data == '?'].count())
auto_data1 = auto_data[(auto_data['normalized-losses'] != '?') & (auto_data['price'] != '?')]
price = auto_data1['price'].astype(float)
norm_loss = auto_data1['normalized-losses'].astype(float)
print(price.corr(norm_loss))  ## no strong linear relation ship between price and normalized-losses
plt.scatter(norm_loss, price) ## no non-linear relationship between price and normalized-losses
plt.xlabel("norm_loss")
plt.ylabel("price")
plt.grid()
plt.show()
auto_data = auto_data.drop('normalized-losses', axis = 1) ## As there is no relationship with price we can drop the normalized-losses as it is having more number of '?' 
```

```python
print(auto_data[auto_data == '?'].count())
print(auto_data.shape)
auto_data = auto_data[(auto_data['num-of-doors'] != '?') & (auto_data['horsepower'] != '?') & 
                                (auto_data['peak-rpm'] != '?') & (auto_data['price'] != '?')]
print(auto_data.shape)
```

```python
# auto_data[['price','peak-rpm','horsepower']].nunique()
# df = df.astype({"Column 1": float, "Column 2": int})
obj_cols = auto_data.select_dtypes(include=['object']).columns.to_list()
print(auto_data[obj_cols].nunique().sort_values(ascending = False))
auto_data = auto_data.astype({"price": float, "horsepower":float})
auto_data.info()
```

```python
## correlation analysis
corr = auto_data.corr()
print(corr[abs(corr['price']) < 0.2]['price'])
ax = sns.heatmap(abs(corr), vmin=0, vmax =1, center = 0, cmap = 'summer')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
## symboling, compression-ratio, height do not have strong linear relationship with price
```
